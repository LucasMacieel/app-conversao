```python
# Importar biblioteca math
import math

def tipo_e_quantidade_laminas(frequencia, potencia_aparente,
                   tensao_primaria, tensao_secundaria):

    # Constantes utilizadas
    Bm = 1.2 # Densidade de fluxo magnético em T
    J = 3 # Densidade de corrente em A/mm2
    Kf = 0.9 # Fator de empilhamento
    Ku = 0.4 # Fator de utilização

    # Calcular a relação de transformação
    a = tensao_primaria/tensao_secundaria

    # Calcular a corrente primária
    # Potência/Tensão
    Ip = potencia_aparente/tensao_primaria
    # Ex: 300Va/120Vp = 2,5A

    # Calcular a corrente secundária
    # OBS: Acrescenta-se à potência secundária 10%
    # de seu valor a fim de se terem em consideração as perdas.
    # Potência/Tensão
    W2 = potencia_aparente * 1.1
    Is = W2/tensao_secundaria
    # Ex: 330Va/220Vp = 1,5A

    # Escolhendo-se a densidade de corrente d = 3 A/mm^2 
    # de acordo com a tabela Potência x Densidade.
    # Seção condutor primário S1
    S1 = Ip/3
    # Ex: 2,5/3 = 0,83 mm^2
    # Obs: livro menciona usar fio número xpto (dá onde ele achou isso ?)

    # Seção condutor secundário S2
    S2 = Is/3
    # Ex: 2,5/3 = 0,5 mm^2
    # Obs: livro menciona usar fio número xpto (dá onde ele achou isso ?)

    # Seção magnética do núcleo
    Sm = 7.5*math.sqrt(W2/frequencia)

    # Seção geométrica
    Sg = 1.1 * Sm

    # Dimensões do núcleo central 5 x 4 cm. (dá onde ?)

    # --------------------------------------------

    # Calcular a área da seção transversal do núcleo
    A = potencia_aparente/(4.44*frequencia*Bm*Kf*Ku)

    # Escolher o tipo de lâmina do núcleo
    # Supondo que o núcleo seja do tipo EI, com as dimensões em cm:
    e = 2.5 # Espessura da coluna
    b = 2.5 # Largura da janela

    # Calcular a quantidade de lâmina do núcleo
    return A*10000/(e*b)


# Mostrar os resultados
# print("Tipo de lâmina do núcleo: EI")
# print("Quantidade de lâmina do núcleo: {:.0f}".format(Q))
```
